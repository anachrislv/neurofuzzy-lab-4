```python
import random
import matplotlib.pyplot as plt
```

```python
def neighbour_exist(i, j):
    if (i, j) in [(1, 1), (2, 2)] or i < 0 or j < 0 or i > 3 or j > 3:
        return False
    return True
```

```python
def neighbours(i, j):
    list_i = []
    list_j = []
    pointer_i = [i, i, i - 1, i + 1]
    pointer_j = [j - 1, j + 1, j, j]
    
    for k in range(4):
        if neighbour_exist(pointer_i[k], pointer_j[k]):
            list_i.append(pointer_i[k])
            list_j.append(pointer_j[k])
    
    return (list_i, list_j)
```

```python
def next_step(i, j):
    n_i, n_j = neighbours(i, j)
    idx = random.randint(0, len(n_i) - 1)
    return n_i[idx], n_j[idx]
```

```python
def robot():
    curr_i, curr_j, steps = 0, 0, 0
    
    while not (curr_i == 2 and curr_j == 3):
        curr_i, curr_j = next_step(curr_i, curr_j)
        steps += 1
    
    return steps
```

```python
simulations = 1
total_steps = 0

for _ in range(simulations):
    steps = robot()
    total_steps += steps
    
average_steps = total_steps / simulations
print(f"Average number of steps per simulation to reach the goal in {simulations} simulation(s): {average_steps:.2f}")
```

```python
#b


simulations = 1000
times = []

for _ in range(simulations):
    steps = robot()
    times.append(steps)

plt.hist(times, bins=range(min(times), max(times) + 1, 1), edgecolor='#FF69B4', color='#FF69B4')
plt.xlabel('Time to Absorption')
plt.ylabel('Frequency')
plt.title(f'Distribution of Time to Absorption for {simulations} Simulations')
plt.show()
```
